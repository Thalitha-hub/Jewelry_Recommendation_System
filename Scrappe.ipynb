```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Read the DataFrame from a CSV file
df = pd.read_csv('group_2.csv')
counter = 0

# Create a session object and authenticate
session = requests.Session()
payload = {
    '_username': 'tobias.braun@camsol.io',
    '_password': '38myt6iBpi'
}
session.post("https://my.the-grow.com/de/login", data=payload)

# Create an empty list to store scraped data
data = []

# Loop through each link in the 'Links' column
for link in df['Link']:
    try:
        # Send a GET request using the authenticated session
        response = session.get(link)
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract relevant data from the soup object
        #company_name = soup.find('h1').text.strip()
        #split_text = company_name.split("aus")
        #location = split_text[1].strip()

        # Extract company, location, and industry
        company_location = soup.find('h1').text.strip().split("aus")
        if len(company_location) == 2:
            company_name = company_location[0].strip()
            location = company_location[1].strip()
        else:
            company_name = "N/A"
            location = "N/A"



        industry = soup.find('span', class_='bg-gray-100 py-1.5 px-4 rounded-full').text.strip() if soup.find('span', class_='bg-gray-100 py-1.5 px-4 rounded-full') else ""

        produktbeschreibung_tag = soup.find('h4', string='Produktbeschreibung')
        produktbeschreibung = produktbeschreibung_tag.find_next('div', class_='text-sm text-justify').get_text(strip=True) if produktbeschreibung_tag else "N/A"

        # Extract vertrieb related information
        vertriebserfolge_tag = soup.find('h4', string='Bisherige Vertriebserfolge')
        vertriebserfolge = vertriebserfolge_tag.find_next('div', class_='text-sm').get_text(strip=True) if vertriebserfolge_tag else "N/A"

        mitbewerber_analyse_tag = soup.find('h4', string='Mitbewerber-Analyse')
        mitbewerber_analyse = mitbewerber_analyse_tag.find_next('div', class_='text-sm').get_text(strip=True) if mitbewerber_analyse_tag else "N/A"

        marktsituation_tag = soup.find('h4', string='Beschreibung der Marktsituation')
        marktsituation = marktsituation_tag.find_next('div', class_='text-sm').get_text(strip=True) if marktsituation_tag else "N/A"

        marktanalyse_tag = soup.find('h3', string='Durchgeführte Marktanalyse')
        marktanalyse = marktanalyse_tag.find_next('div', class_='text-primary').get_text(strip=True) if marktanalyse_tag else "N/A"

        zielbranchen_tag = soup.find('h3', string='Zielbranchen')
        zielbranchen = zielbranchen_tag.find_next('div', class_='text-primary').get_text(strip=True) if zielbranchen_tag else "N/A"

        sales_channel_tag = soup.find('h3', string='Sales-Channel')
        sales_channel = sales_channel_tag.find_next('div', class_='text-primary').get_text(strip=True) if sales_channel_tag else "N/A"

        vertriebsaktivitaeten_tag = soup.find('h3', string='Vertriebsaktivitäten')
        vertriebsaktivitaeten = vertriebsaktivitaeten_tag.find_next('div', class_='text-primary').get_text(strip=True) if vertriebsaktivitaeten_tag else "N/A"

        team_members = [member.find('div', class_='text-center').get_text(strip=True) for member in soup.find_all('a', class_='btn-person')]
        members_str = ", ".join(team_members)

        finanzierungs_summe_tag = soup.find('h3', string='Finanzierungs-Summe')
        finanzierungs_summe = finanzierungs_summe_tag.find_next('div', class_='text-primary').get_text(strip=True) if finanzierungs_summe_tag else "N/A"

        phase_tag = soup.find('h3', string='Phase')
        phase = phase_tag.find_next('div', class_='text-primary').get_text(strip=True) if phase_tag else "N/A"

        art_der_finanzierung_tag = soup.find('h3', string='Art der Finanzierung')
        art_der_finanzierung = art_der_finanzierung_tag.find_next('div', class_='text-primary').get_text(strip=True) if art_der_finanzierung_tag else "N/A"

        finanzierungsziel_tag = soup.find('h3', string='Finanzierungsziel')
        finanzierungsziel = finanzierungsziel_tag.find_next('div', class_='text-primary').get_text(strip=True) if finanzierungsziel_tag else "N/A"

        business_type_tag = soup.find('h3', string='Business type')
        business_type = business_type_tag.find_next('div', class_='text-blue-500 mt-1').get_text(strip=True) if business_type_tag else "N/A"

        looking_for_investment_tag = soup.find('h3', string='Looking for investment')
        looking_for_investment = looking_for_investment_tag.find_next('div', class_='text-blue-500 mt-1').get_text(strip=True) if looking_for_investment_tag else "N/A"

        scraped_data = {
            "Company Name": company_name,
            "Location": location,
            "Industry": industry,
            "Produktbeschreibung": produktbeschreibung,
            "Vertriebserfolge": vertriebserfolge,
            "Mitbewerber-Analyse": mitbewerber_analyse,
            "Beschreibung der Marktsituation": marktsituation,
            "Durchgeführte Marktanalyse": marktanalyse,
            "Zielbranchen": zielbranchen,
            "Sales-Channel": sales_channel,
            "Vertriebsaktivitäten": vertriebsaktivitaeten,
            "Team Members": members_str,
            "Finanzierungs-Summe": finanzierungs_summe,
            "Phase": phase,
            "Art der Finanzierung": art_der_finanzierung,
            "Finanzierungsziel": finanzierungsziel,
            "Business Type": business_type,
            "Looking for Investment": looking_for_investment
        }

        data.append(scraped_data)
        #counter += 1

    except Exception as e:
        print(f"Error occurred while scraping link {link}: {str(e)}")

scraped_df = pd.DataFrame(data)
#save the data to csv
```
